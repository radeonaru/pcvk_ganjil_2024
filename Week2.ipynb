**Jawaban**

1. Perbedaan antara gambar yang ditampilkan dengan dan tanpa Matplotlib terletak pada tingkat visualisasi dan kontrol yang diberikan dalam penyajian data. Ketika gambar ditampilkan tanpa menggunakan Matplotlib, biasanya data hanya ditampilkan dalam bentuk dasar seperti teks atau representasi visual sederhana tanpa kemampuan untuk memberikan informasi grafis yang mendetail. Sedangkan ketika menggunakan Matplotlib, gambar akan ditampilkan dalam plot yang memungkinkan kontrol lebih banyak, seperti zoom, pan, dan penambahan anotasi. Matplotlib juga memudahkan penambahan elemen lain seperti grafik dan label, serta menyimpan gambar dalam berbagai format dengan pengaturan kualitas.

2. Perbedaan utama antara tipe data int16 dan int32 dalam pembuatan gambar hitam terletak pada rentang nilai dan kebutuhan memori. int16 menggunakan 16 bit per piksel, dapat menyimpan nilai dari -32,768 hingga 32,767, dan memerlukan 2 byte per piksel. sedangkan int32 menggunakan 32 bit per piksel, dapat menyimpan nilai dari -2,147,483,648 hingga 2,147,483,647, dan memerlukan 4 byte per piksel. Untuk gambar hitam, keduanya akan memiliki piksel dengan nilai 0, tetapi int32 memerlukan lebih banyak memori dibandingkan int16.

3. Potongan kode tersebut digunakan untuk menampilkan gambar di Google Colab. Karena metode biasa untuk menampilkan gambar dari OpenCV tidak bekerja di Colab, sedangkan fungsi cv2_imshow digunakan untuk melihat gambar yang diolah langsung di notebook Colab.

4. io dari scikit-image menyediakan fungsi untuk membaca gambar dari file atau URL. Ini berguna untuk memuat gambar ke dalam program Anda agar bisa diproses lebih lanjut. selain itu, skimage.io mendukung berbagai format gambar seperti JPEG, PNG, TIFF, dan lainnya. Dengan skimage.io, Anda dapat memuat gambar dari sumber eksternal sehingga bisa diproses atau ditampilkan di Google Colab
